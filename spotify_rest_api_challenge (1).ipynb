### Set your main variables!!!

Set the `base_url` (i.e.: End-point) and `headers` for your __GET__ operations

```python
# setting the main variables

base_url = 'https://api.spotify.com/v1'
resource = '/artists/'
artist_id = '711MCceyCBcFnzjGY4Q7Un'
related_artists = '/related-artists'
top_tracks = '/top-tracks'
market = '?market=ES'
headers = {'Authorization':'Bearer {token}'.format(token=access_token)}
```

```python
# your code

response = requests.get(base_url + resource + artist_id + related_artists, headers=headers).json()
df_related_artists = pd.DataFrame(response['artists'])
df_related_artists = df_related_artists
df_related_artists
```

```python
response2 = requests.get(base_url + resource + artist_id + top_tracks + market, headers=headers).json()
df_top_tracks = pd.DataFrame(response2['tracks'])[['artists','id','name']]
df_top_tracks
```

```python
top_tracks_per_artist = df_top_tracks.copy()

id_list = []
for i in df_related_artists['id']:
    id_list.append(i)
    related_art_top_tracks = requests.get(base_url + resource + i + top_tracks + market, headers=headers).json()
    df_obj = pd.DataFrame(related_art_top_tracks['tracks'])[['artists','id','name']]
    top_tracks_per_artist = top_tracks_per_artist.append(df_obj, ignore_index=True)

display(top_tracks_per_artist)


'''
The result showed 210 records because there were 20 related artist and the original one (AC/DC) is also included
'''
```
