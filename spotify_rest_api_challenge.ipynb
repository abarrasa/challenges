```python
# imports
import pandas as pd
import numpy as np
import requests
```

### Get access!!!

Get your `client_id` and `client_secret` a generate your __token__ access

Hay que rellenar las variables.
Para crear el token, hay que hacer una peticion post a spotify para que te devuelva el token pero hay que pedir unos permisos. 'auth_url' es el 'end point'.
En 'auth_response' ponemos el end point y varios headers, y el servidor nos devuelve el token.

```python
# generate token with a POST request

client_id = '8c5be5d8e64244c8ad0c8274c85fcc7e'
client_secret = '0c396ca53d2448bdace07a4ac51b4146'
auth_url = 'https://accounts.spotify.com/api/token'

auth_response = requests.post(auth_url, {'grant_type': 'client_credentials',
                                         'client_id': client_id,
                                         'client_secret': client_secret}).json()
access_token = auth_response['access_token']
```

### Set your main variables!!!

Set the `base_url` (i.e.: End-point) and `headers` for your __GET__ operations

```python
# setting the main variables
base_url = 'https://api.spotify.com/v1'
resource = '/artists'
artist_id = '/711MCceyCBcFnzjGY4Q7Un'
related_artists = '/related-artists'
headers = {'Authorization':'Bearer {token}'.format(token=access_token)}
top_tracks = '/top-tracks'
country = '?market=ES'
```
